```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_neural_network(input_dim, output_dim, hidden_dims):
    # Initialize graph
    G = nx.DiGraph()
    
    # Define layers
    layers = [input_dim] + hidden_dims + [output_dim]
    
    # Create nodes for each layer
    for i, layer_size in enumerate(layers):
        for j in range(layer_size):
            G.add_node(f"L{i}_{j}", pos=(i, -j))
    
    # Create edges between layers
    for i in range(len(layers) - 1):
        for j in range(layers[i]):
            for k in range(layers[i+1]):
                G.add_edge(f"L{i}_{j}", f"L{i+1}_{k}")
    
    pos = nx.get_node_attributes(G, 'pos')
    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=False, node_size=500, node_color='lightblue', edge_color='gray')
    plt.title('4-Layer Fully Connected Neural Network')
    plt.show()

# Define network dimensions
input_dim = 4
output_dim = 88
hidden_dims = [30, 30, 30]

draw_neural_network(input_dim, output_dim, hidden_dims)
```
